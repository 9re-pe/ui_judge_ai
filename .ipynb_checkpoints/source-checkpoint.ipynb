```python
# 設問データ(TSVファイル出力)

# データ生成
url = 'https://www.u.tsukuba.ac.jp/~s2010780/experimentA/image/{}.png'
qe = '直感で「クリックしたいと思った」ボタンを選んでください'
msg = '提示されているボタンをクリックして選択するのではなく、ボタン画像の下にある白いボタンをクリックして選択してください。'
tsv_data = []
for i, v in enumerate(battle_table_shuffled):
    row = []
    # 設問ID(半角英数字20文字以内)
    row.append(i)
    # チェック設問有無(0:無 1:有)
    row.append(0)
    # チェック設問の解答(F04用)
    row.append('')
    # チェック設問の解答(F07用)
    row.append('')
    # F01:ラベル
    row.append(qe)
    # F02:ラベル
    row.append('ボタン1')
    # F03:画像（ＵＲＬ）
    row.append(url.format(v[0]))	
    # F04:選択ボタン
    row.append('ボタン1')
    # F05:ラベル
    row.append('ボタン2')
    # F06:画像（ＵＲＬ）
    row.append(url.format(v[1]))
    # F07:選択ボタン
    row.append('ボタン2')
    # F08:ラベル
    row.append(msg)
    
    tsv_data.append(row)


# データ出力
import csv
f = open('out.tsv', 'w', encoding='utf-8', newline='')
writer = csv.writer(f, delimiter='\t')
writer.writerows(tsv_data)
f.close()
```

```python
# 納品データの受け取り
# resules = [(ボタンNo, ボタンNo, ワーカ1の解答, ワーカ2の解答)]
results = []

f = open('result.txt', 'r')
i = 0
while True:
    line = f.readline()
    if line == '':
        break
    line = str(line).strip()
    line_li = line.split(',')
    if i % 2 == 0:
        tmp = line_li[3]
    else:
        results.append((int(line_li[1]), int(line_li[2]), int(tmp), int(line_li[3])))
    i += 1
f.close()
```

```python
# デバッグ
for i in results:
    print(i)
```

```python
# ワーカーを測る
qe_num = len(results)
same_num = 0
for result in results:
    if result[2] == result[3]:
        same_num += 1
same_rate = same_num / qe_num * 100

# 出力
print('問題数(種類): {}'.format(qe_num))
print('ふたりのワーカーの解答が一致していた問題数: {}'.format(same_num))
print('ワーカー間の解答一致率: {}%'.format(same_rate))
```

```python
# データにラベルをつける

# 各ボタンの持ち点を算出
points = [0] * BUTTON_NUM # ボタンNoがインデックスと対応
for result in results:
    # ワーカー1
    if result[2] == 1:
        points[result[0]] += 1
        points[result[1]] -= 1
    else:
        points[result[0]] -= 1
        points[result[1]] += 1
    # ワーカー2
    if result[3] == 1:
        points[result[0]] += 1
        points[result[1]] -= 1
    else:
        points[result[0]] -= 1
        points[result[1]] += 1

# 持ち点をラベルに持つデータ
data_point = []
for li in data_pre:
    data_point.append([li[0], li[1], li[2], li[3], li[4], points[li[0]]])

# 押したいか押したくないかをラベルの持つデータ
data_2labels = []
for li in data_pre:
    if points[li[0]] > 0:
        data_2labels.append([li[0], li[1], li[2], li[3], li[4], 1])
    else:
        data_2labels.append([li[0], li[1], li[2], li[3], li[4], 0])
```

```python
# 機械学習
def ml(data, k):
    # データフレーム化
    import pandas as pd
    columns =['button_no', 'brightness', 'saturation', 'roundness', 'popout', 'class']
    df = pd.DataFrame(data = data, columns = columns)

    # データをシャッフル
    df = df.sample(frac=1).reset_index(drop=True)

    # データの切り分け
    # 教師 : テスト = 8 : 2
    train_rate = int(df.shape[0] * 0.8)
    train_data = df.iloc[:train_rate, :]
    test_data = df.iloc[train_rate:, :]
    test_button_no = test_data.iloc[:, 0].values

    # パラメータがどの列か指定
    params = list(range(df.shape[1]))[1:-1]
    
    # knn
    from sklearn.neighbors import KNeighborsClassifier as KNN
    knn = KNN(n_neighbors = k, n_jobs = -1)
    # 教師データセット
    knn.fit(
        X = train_data.iloc[:, params].values,
        y = train_data["class"].values,
    )
    #テストデータをクラス分類
    predict = knn.predict(
        X = test_data.iloc[:, params].values, 
    )
    results = []
    for i in range(len(predict)):
        results.append([test_button_no[i], predict[i]])
    # 認識率
    acc = sum(test_data["class"] == predict) / len(predict) * 100

    return [results, acc]
```
